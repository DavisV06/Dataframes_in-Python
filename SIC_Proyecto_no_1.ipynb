```python
def sexCount(dtFr):
  DT_women = dtFr[dtFr['sex']=='F']
  Women_count = len(DT_women)
  DT_men = dtFr[dtFr['sex']=='M']
  Men_count = len(DT_men)

  return Women_count, Men_count
```

```python
GP = pd.concat([GP_mat,GP_por])
GP

MS = pd.concat([MS_mat,MS_por])
MS
```

```python
def GraficaHM(datFRM, nombre, curso):

  WomenM, manM = sexCount(datFRM)

  labels= ['Hombres','Mujeres']

  colors=['blue', 'red']

  sizes= [manM, WomenM]

  plt.title('Porcentaje de hombres y mujeres '+ nombre+ ' en el curso de '+curso)

  plt.pie(sizes,labels=labels, colors=colors, startangle=90, autopct='%1.2f%%')

  plt.axis('equal')

  plt.show()
```

```python
def graficaEdades(DATFR, nombre, color):
  DATFR_ed=[]
  DATFR_ed = list(DATFR['age'])
  DATFR_ed.sort()
    
  DATFR_ed_disp = []
  DATFR_ed_cont = []
  [DATFR_ed_disp.append(edad) for edad in DATFR_ed if edad not in DATFR_ed_disp]

  DATFR_ed_disp.sort()


  for i in range(len(DATFR_ed_disp)):
    DATFR_ed_cont.append(0)

  for edad in range(len(DATFR_ed_disp)):
    for i in range(len(DATFR_ed)):
      if(DATFR_ed[i]== DATFR_ed_disp[edad]):
        DATFR_ed_cont[edad] += 1

  #GRAFICA DE BARRAS DE EDADES DE GP

  fig = plt.figure(figsize=(9,6))
  ax1 = plt.subplot(111)
  ax1.barh(DATFR_ed_disp, DATFR_ed_cont, color = color)

  plt.title('Cantidad de estudiantes que tienen la misma edad '+ nombre)
  ax1.set_ylabel('Edades')
  ax1.set_xlabel('Cantidades')

  #linea vertical de separacion
  ax1.xaxis.grid(linestyle = '--', linewidth = 0.5)

  plt.show()

```

```python
def promedioED(datFr):

  datFr_edad = list(datFr['age'])

  den = len(datFr_edad)
  num = 0 

  for edades in range(den):
    num = num + datFr_edad[edades]
  
  dat_Fr_prom = num/den

  return dat_Fr_prom
```

```python
#Muestre el promedio de las edades de cada curso de cada escuela
GP_mat_prom = promedioED(GP_mat)
GP_por_prom = promedioED(GP_por)
MS_mat_prom = promedioED(MS_mat)
MS_por_prom = promedioED(MS_por)
print("El promedio de las edades en el curso de matematicas de Gabriel de Pereira es: {:0.2f}\n".format(GP_mat_prom))
print("El promedio de las edades en el curso de portugues de Gabriel de Pereira es: {:0.2f}\n".format(GP_por_prom))
print("El promedio de las edades en el curso de matematicas de Mousinho da Silveira es: {:0.2f}\n".format(MS_mat_prom))
print("El promedio de las edades en el curso de portugues de Mousinho da Silveira es: {:0.2f}\n".format(MS_por_prom))
```

```python
def promNotas(datFr):
  datFr1_nt_mat_lista = list(datFr['G1'])
  datFr2_nt_mat_lista = list(datFr['G2'])
  datFr3_nt_mat_lista = list(datFr['G3'])

  #PROMEDIO DE G1 EN MATEMATICAS 
  num = 0

  for nota in range(len(datFr1_nt_mat_lista)):
    num = num + datFr1_nt_mat_lista[nota]

  prom1 = num/len(datFr1_nt_mat_lista)

  #PROMEDIO DE G2 EN MATEMATICAS
  num = 0

  for nota in range(len(datFr2_nt_mat_lista)):
    num = num + datFr2_nt_mat_lista[nota]

  prom2 = num/len(datFr2_nt_mat_lista)

  #PROMEDIO DE G3 EN MATEMATICAS
  num = 0

  for nota in range(len(datFr3_nt_mat_lista)):
    num = num + datFr3_nt_mat_lista[nota]

  prom3 = num/len(datFr3_nt_mat_lista)

  prom =  (prom1 + prom2 + prom3 )/3

  return prom
```

```python
promGPM = promNotas(GP_mat)
promGPP = promNotas(GP_por)
promMSM = promNotas(MS_mat)
promMSP = promNotas(MS_por)
print("El promedio de las notas G1, G2, G3  en el curso de matematicas en Gabriel de Pereira es: {:0.2f}\n".format(promGPM))
print("El promedio de las notas G1, G2, G3  en el curso de portugues en Gabriel de Pereira es: {:0.2f}\n".format(promGPP))
print("El promedio de las notas G1, G2, G3  en el curso de matematicas en Mousinho da Silveira es: {:0.2f}\n".format(promMSM))
print("El promedio de las notas G1, G2, G3  en el curso de portugues en Gabriel de Pereira es: {:0.2f}\n".format(promMSP))
```

```python
def obtenerPorAsis(asistencias, asist):
  porcentaje = []
  for i in asistencias:
    por = ((asist - i)/asist)*100
    #porst = str(por) + '%'
    porcentaje.append(por)
  
  return porcentaje
```

```python

#TOTAL DE CLASES DE MATEMATICAS EN GP
GPM_class = GP_mat['absences'].max()
GPM_as = list(GP_mat['absences'])
GPM_porc = []
GPM_porc = obtenerPorAsis(GPM_as, GPM_class)

#TOTAL DE CLASES DE PORTUGUES EN GP
GPP_class = GP_por['absences'].max()
GPP_as = list(GP_por['absences'])
GPP_porc = []
GPP_porc = obtenerPorAsis(GPP_as, GPP_class)

#TOTAL DE CLASES DE MATEMATICAS EN MS
MSM_class = MS_mat['absences'].max()
MSM_as = list(MS_mat['absences'])
MSM_porc = []
MSM_porc = obtenerPorAsis(MSM_as, MSM_class)

#TOTAL DE CLASES DE PORTUGUES EN MS
MSP_class = MS_por['absences'].max()
MSP_as = list(MS_por['absences'])
MSP_porc = []
MSP_porc = obtenerPorAsis(MSP_as, MSP_class)

```

```python
def rellenarColum(datFr, porcentaje):
  datFr.loc[:, 'porcentaje'] = porcentaje

  #Si el estudiante tiene un porcentaje de asistencia menor al 80% del curso reprueba 
  datFr.loc[(datFr.porcentaje<=80), 'approved'] = 0

  #Si el estudiante tiene un porcentaje de asistencia mayor al 80% del curso, pero la nota G3 es menor a 10 reprueba
  datFr.loc[(datFr.porcentaje>80)&(datFr.G3<10), 'approved'] = 0

  #Si el estudiante tiene un porcentaje de asistencia mayor al 80% del curso, pero la nota G3 está entre 10 y 15 aprueba y se asigna 1 a la columna extra
  datFr.loc[(datFr.porcentaje>80)&((datFr.G3>=10)&(datFr.G3<=15)), 'approved'] = 1
  datFr.loc[(datFr.porcentaje>80)&((datFr.G3>=10)&(datFr.G3<=15)), 'extra'] = 0

  #Si el estudiante tiene un porcentaje de asistencia mayor al 80% del curso y la nota G3 es mayor a 15 aprueba y se asigna 0 a la columna extra
  datFr.loc[(datFr.porcentaje>80)&(datFr.G3>=15),'approved'] = 1
  datFr.loc[(datFr.porcentaje>80)&(datFr.G3>=15),'extra'] = 1

  datFRN = datFr.drop(['porcentaje'], axis = 1)

  return datFRN
```

```python
#CREE UNA NUEVA COLUMNA LLAMADA EXTRA
GP_mat["extra"] = ""
GP_por["extra"] = ""
MS_mat["extra"] = ""
MS_por["extra"] = ""

#CREE UNA NUEVA COLUMNA LLAMADA APPROVED
GP_mat["approved"] = ""
GP_por["approved"] = ""
MS_mat["approved"] = ""
MS_por["approved"] = ""

GP_mat = rellenarColum(GP_mat, GPM_porc)
GP_por = rellenarColum(GP_por, GPP_porc)
MS_mat = rellenarColum(MS_mat, MSM_porc)
MS_por = rellenarColum(MS_por, MSP_porc)
```

```python
def conteoAprob(datFr):
  a = datFr[datFr['approved']==1]
  aprobados = len(a)
  r = datFr[datFr['approved']==0]
  reprobados = len(r)
  
  return aprobados, reprobados
```

```python
MAT = pd.concat([GP_mat,MS_mat])
MAT_ap, MAT_rp = conteoAprob(MAT)

POR = pd.concat([GP_por,MS_por])
POR_ap, POR_rp = conteoAprob(POR)
```
